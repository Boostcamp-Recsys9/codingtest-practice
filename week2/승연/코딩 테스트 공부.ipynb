```python
import math
# 모든 문제를 풀기 위한 최대 점수 m,n
# 풀 수 있는 문제들 중 부족한 부분을 가장 많이 채워줄 수 있는 과목을 선택
def check_fast_resolve(alp, cop, problems):
    t = []
    for v in problems:
        t.append([v[0] - alp, v[1] - cop])
    req_min_idx = t.index(min(t))
    return problems[req_min_idx]


def add_one_hour(alp, cop, req_min):
    answer = 0
    while alp < req_min[0] or cop < req_min[1]:
        if alp < req_min[0]:
            alp+=1
            answer+=1

        if cop < req_min[1]:
            cop+=1
            answer+=1    
        
    return alp, cop, answer
    

def solution(alp, cop, problems):
    answer = 0
    
    alg_max = [i[0] for i in problems]
    cod_max = [i[1] for i in problems]
        
    # 풀 수 있는 최대치
    req_max = [max(alg_max), max(cod_max)]
    
    # 초기 시점으로부터 가장 빨리 풀 수 있는 문제
    req_min = check_fast_resolve(alp, cop, problems)
    
    
    # 풀 수 있는 문제들
    can_resolve = [i for i in problems if alp>=i[0] and cop>=i[1]]
    
    
    # 알고리즘에서 부족한 부분, 코드에서 부족한 부분을 가장 많이 채워줄 수 있는지
    # case1: 부족한 하나를 채우고 다른 부족한걸 다음에 채우는 경우
    # case2: 한번에 여러개를 채울 수 있는 경우
    
    # 최소를 만족하지 못할 경우 최소 can_resolve하나를 구해주기 위함
    if alp < req_min[0] or cop < req_min[1]:
        alp, cop, answer = add_one_hour(alp, cop, answer, req_min)
        can_resolve.append(req_min)
    
    while True:
        if alp >= req_max[0] and cop >= req_max[1]:
            break

        #현재 alg,cod에서 더해주는 벡터지점을 임시로 담아둘 공간
        for idx, v in enumerate(problems):
            if alp >= v[0] and cop >= v[1]:
                if not problems[idx] in can_resolve:
                    can_resolve.append(problems[idx])

        print("can_resolve : ", can_resolve)
        
        #문제 풀이
        t_vect = []
        for i, p in enumerate(can_resolve):
            t_vect.append(math.sqrt(((alp + p[2])**2 + (cop + p[3])**2)))
            if p in problems:
                problems.pop(problems.index(p))
        t_idx = t_vect.index(max(t_vect))
            
        req_min2 = check_fast_resolve(alp, cop, problems)
        print("req_min2 : ", req_min2)
        print("alp, cop, answer : ", alp, cop, answer)
        t_alp, t_cop, t_answer = add_one_hour(alp, cop, req_min2)

        if can_resolve[t_idx][4] < t_answer:
            print('aa')
            print(can_resolve[t_idx][4], t_answer)
            alp += can_resolve[t_idx][2]
            cop += can_resolve[t_idx][3]
            answer += can_resolve[t_idx][4]
            print("alp, cop, answer : ", alp, cop, answer)
        else:
            print('bb')
            print(can_resolve[t_idx][4], t_answer)
            alp += t_alp
            cop += t_cop
            answer += t_answer
            
        print('-------')
    
    
    return answer
```
